```python
from pylab import plt, mpl
import numpy as np
import seaborn as sns

plt.style.use('seaborn')
mpl.rcParams['savefig.dpi'] = 500
mpl.rcParams['font.family'] = 'Times new roman'
```

```python
proposed = [0.76, 0.90, 0.82]
lppls = [0.68, 0.82, 0.74]
psy = [0.66, 0.82, 0.73]

proposed1 = [0.76, 0.90, 0.82]
lppls1 = [0.68, 0.82, 0.74]
psy1 = [0.66, 0.82, 0.73]

# 设置柱状图的宽度
bar_width = 0.25

# 设置x轴标签位置
r1 = np.arange(len(proposed))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]

sns.set_palette("pastel")

# 绘制柱状图
bar1 = plt.bar(r1, proposed, color=sns.color_palette()[0], width=bar_width, edgecolor='black', label='Proposed', hatch = '/')
bar2 = plt.bar(r2, lppls, color=sns.color_palette()[1], width=bar_width, edgecolor='black', label='LPPLS', hatch = '//')
bar3 = plt.bar(r3, psy, color=sns.color_palette()[2], width=bar_width, edgecolor='black', label='PSY')

# 添加数据标签
for bar in bar1+bar2+bar3:
    plt.text(bar.get_x() + bar.get_width()/2., bar.get_height(), '%.2f' % bar.get_height(), ha='center', va='bottom')

# 添加x轴标签
plt.xlabel('Performance Metrics', fontweight='bold')
plt.xticks([r + bar_width for r in range(len(proposed))], ['Precision ', 'Recall', 'F1-score'])
plt.ylim(0.6,1)

# 添加y轴标签
plt.ylabel('Scores', fontweight='bold')
#plt.grid()

# 添加图例
plt.legend()
plt.savefig('Fig-Comparation.svg', dpi = 500, format='svg')
plt.show()
```
