```python
import pandas as pd
import numpy as np
#import matplotlib.pyplot as plt
#import matplotlib
from pylab import plt, mpl
mpl.rcParams['savefig.dpi'] = 600
mpl.rcParams['font.family'] = 'Times New Roman'
%config InlineBackend.figure_format = 'svg'
```

```python
df = pd.read_csv('aiif_eikon_eod_data.csv').set_index('Date')
df.head()
```

```python
plt.style.use('seaborn-ticks')
plt.figure(figsize=(9,3))
plt.plot(df['GLD'].values,c = 'k')
plt.grid('True',linestyle = '-')
ax = plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.xlim(0,2600)
plt.ylabel('Price',fontname = 'Times new roman')
plt.xlabel('Time',fontname = 'Times new roman')
plt.title('Gold price',fontname = 'Times new roman')
```
